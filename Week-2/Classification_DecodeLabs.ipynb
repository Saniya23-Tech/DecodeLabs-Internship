```python
#Supervised Machine Learning Classification on the Iris Dataset
#import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, f1_score
```

```python
#step-01: Load Iris Dataset
iris=load_iris()
X= iris.data #Features
Y=iris.target #Labels
print("Iris Dataset Loaded Successfully")
print("Number of samples:", X.shape[0])
print("Number of Features:", X.shape[1])
print("Target Classes:", iris.target_names)
```

```python
#step-02: Feature scaling(For the distance for KNN )
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X) # x-mean/SD
print("\n Feature Scaling Completed")
```

```python
#step-03: Train-Test Split (Dividing the data sets into training and testing)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y , test_size=0.2, random_state=42) # we are testing the 20% and training the 80% to learn the patterns not only memorize 
print("\n Train_Test Split Completed")
print("Training Samples:", len(X_train))
print("Testing Samples:", len(X_test))
```

```python
#step-04: Apply KNN
knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train) #Training the data and labels
print("\n KNN Model Training Completed")
```

```python
#step-05: Prediction (testing on unseen data)
y_pred = knn.predict(X_test) #testing i/p data 
print("\n Predictions Generated")
```

```python
#step-06: Evaluation
cm=confusion_matrix(Y_test , y_pred)
f1=f1_score(Y_test, y_pred, average='weighted')
print("\n Confusion Matrix:" , cm )
print("\n F1 Score:" , round(f1,4))
```
