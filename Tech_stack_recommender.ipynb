```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("careers.csv")
print("Dataset loaded successfully")
```

```python
user_skills=input("Enter your skills:")
print("Skills You Have:", user_skills)
```

```python
all_skills=df["Skills"].tolist()
all_skills.append(user_skills)
print(all_skills)

```

```python
vectorizer= TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(all_skills)
print("TF-IDF Vectorization Completed!")
```

```python
#The negative indexing is used to access elements from an the end of a list/array
user_vector = tfidf_matrix[-1] # row 3 (your skills)
career_vectors = tfidf_matrix[:-1] # row 1 and 2 careers
```

```python
similarity_scores = cosine_similarity(user_vector, career_vectors)
similarity_scores = similarity_scores.flatten()
print(similarity_scores)
```

```python
df["Similarity Score"] = similarity_scores

```

```python
recommended = df.sort_values(
    by="Similarity Score",
    ascending=False
)
```

```python
top_3 = recommended.head(3)

print("\n======================================")
print("Top 3 Recommended Career Paths")
print("======================================\n")

for index, row in top_3.iterrows():
    print(f"{row['career']} --> Score: {row['Similarity Score']:.3f}")
```
